```python
mice = expert_dlc_hdf.columns.levels[1]
bodyparts = expert_dlc_hdf.columns.levels[2]

coord_dict = {}

for bp in bodyparts:
    dist = 0
    for mouse in mice:

        x_Expert = expert_dlc_hdf['', mouse, bp, 'x'].values
        y_Expert = expert_dlc_hdf['', mouse, bp, 'y'].values

        x_inExpert = inExpert_dlc_hdf['', mouse, bp, 'x'].values
        y_inExpert = inExpert_dlc_hdf['', mouse, bp, 'y'].values
        
        # sum_x += np.sum(abs(x_Expert - x_inExpert))    
        # dist += np.sqrt((x_Expert - x_inExpert)**2 + (y_Expert - y_inExpert)**2)
        dist += np.nansum(np.sqrt((x_Expert - x_inExpert)**2 + (y_Expert - y_inExpert)**2))

    # error = sum_x / (len(expert_dlc_hdf) * len(mice))
    error = dist / (len(expert_dlc_hdf) * len(mice))

    coord_dict.update({bp:error})
    
    
```

```python
coord_dict
```

```python

with open('MAE_bp_2users.csv', 'w') as f:  # You will need 'wb' mode in Python 2.x
    w = csv.DictWriter(f, coord_dict.keys())
    w.writeheader()
    w.writerow(coord_dict)
```
